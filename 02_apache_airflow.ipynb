# *Apache Airflow*.

[*Apache Airflow*](https://airflow.apache.org/) es una plataforma de código abierto especializada para desarrollar, programar y monitorear flujos de trabajo (*workflows*) por lotes (*batch*). 

Debido a que *Apache Airflow* es capaz de importar y aprovechar las diversas bibliotecas basadas en *Python*, permite crear flujos de trabajo que se conectan con prácticamente cualquier tecnología. 

Cuenta con una interfaz web que permite administrar el estado de sus flujos de trabajo. 

*Apache Airflow* se puede implementar desde un solo proceso en una computadora personal hasta una configuración distribuida.

La documentación de *Apache Airflow* puede ser consultada desde:

* https://airflow.apache.org/docs/apache-airflow/stable/index.html

### Instalación de *Apache Airflow*.

*Apache Airflow* puede ser instalado utilizando *pip*, como un paquete de *Helm* para *Kubernetes* e incluso puede ser utilizado como un servicio administrado como es el caso de [*Cloud Composer*](https://cloud.google.com/composer/docs).

Los componentes básicos de *Apache Airflow* están contenidos en el paquete  [```apache-airflow```](https://pypi.org/project/apache-airflow/).

Entre los [requisitos](https://airflow.apache.org/docs/apache-airflow/stable/installation/prerequisites.html) de instalación es necesario contar con una conexión a una base de datos compatible. 

**Nota:** Para los entornos de prueba en computadoras personales, se recomienda el uso de *SQLite*.

````python
pip install apache-airflow
````

###  Los *providers* de *Apache Airflow*.

Los *providers* de *Apache Airflow* son paquetes que extienden la funcionalidades básicas del paquete ```apache-airflow``` de tal forma que pueda interactuar con diversas platafomas de nube, bases de datos y otras herramientas.

El listado de *providers* de *Apache Airflow* puede ser consultado desde:

https://airflow.apache.org/docs/apache-airflow-providers/packages-ref.html

## Los *DAG* en *apache Airflow*.

Los flujos de trabajo de *Apache Airflow* se comportan como [gráficos acíclicos dirigidos (*DAG*)](https://es.wikipedia.org/wiki/Grafo_ac%C3%ADclico_dirigido) compuestos por tareas (*tasks*). 

El programador de Airflow ejecuta sus tareas en una matriz de trabajadores mientras sigue las dependencias especificadas. Las ricas utilidades de la línea de comandos facilitan la realización de cirugías complejas en DAG. La interfaz de usuario enriquecida facilita la visualización de canalizaciones que se ejecutan en producción, supervisa el progreso y soluciona problemas cuando es necesario.



## La arquitectura de *Apache Airflow*.

Una implementación de *Apache Airflow* consiste principalmente en:

* Un *scheduler*, que coordina tanto la activación de los flujos de trabajo como el envío de tareas a un ejecutor (*executor*) para que las realice.
* Un cluster de ejecutores (*executor*) que gestionan las tareas (*tasks*). En la instalación básica, el *scheduler* y el ejecutor están en la misma máquina, pero en entornos de producción,los ejecutores se distribuyen entre varios equipos de cómputo a los que se les llama *workers*.
* Un servidor web, con una interfaz de usuario que permite inspeccionar, activar y depurar el comportamiento de los *DAG* y las tareas.
* Una *data store* de *DAG*, la cual es consultada por el *scheduler* y los diversos ejecutores.
* Una base de datos de metadatos, utilizada por el *scheduler*, el ejecutor y el servidor web para almacenar el estado del sistema.



Los detalles de la arquitectura de *Airflow* pueden ser consutados desde:

* https://airflow.apache.org/docs/apache-airflow/stable/core-concepts/overview.html
